```python
from tkinter import *
import qrcode
from PIL import Image,ImageTk
from resizeimage import resizeimage

class Qr_generator:
    def __init__(self, root):
        self.root = root
        self.root.geometry('900x650+200+50')
        self.root.title('QR code generation | developed by Ashwini')
        self.root.resizable(False, False)

        title = Label(self.root, text='Qr Code Generation', font=("times new roman", 40), bg='#053246',
                      fg='white').place(x=0, y=0, relwidth=1)

        # ************************* Employee Details Window **********************************
        # ********** variables ************************

        self.var_emp_code=StringVar()
        self.var_emp_name = StringVar()
        self.var_emp_department = StringVar()
        self.var_emp_designation = StringVar()
        self.var_emp_email=StringVar()
        # self.var_emp_gender = StringVar()
        self.var_emp_addr = StringVar()
        self.var_emp_sal=StringVar()


        emp_frame = Frame(self.root, bd=2, relief=RIDGE, bg='white')
        emp_frame.place(x=50, y=160, width=500, height=480)

        emp_title = Label(emp_frame, text='Employee Details', font=("Goudy old Style ", 20), bg='#043256',fg='white').place(x=0, y=0, relwidth=1)

        lbl_emp_code = Label(emp_frame, text='Employee ID', font=('times new roman', 15, 'bold'), bg='white').place(x=20, y=60)
        lbl_emp_name = Label(emp_frame, text='Employee Name', font=('times new roman', 15, 'bold'), bg='white').place(x=20, y=100)
        lbl_department = Label(emp_frame, text='Department', font=('times new roman', 15, 'bold'), bg='white').place(x=20, y=140)
        lbl_designation = Label(emp_frame, text='Designation', font=('times new roman', 15, 'bold'), bg='white').place(x=20, y=180)
        lbl_emp_email = Label(emp_frame, text='E-mail', font=('times new roman', 15, 'bold'), bg='white').place(x=20,y=220)
        lbl_emp_gender = Label(emp_frame, text='Emp Gender', font=('times new roman', 15, 'bold'), bg='white').place(x=20, y=260)
        lbl_addr = Label(emp_frame, text='Emp Address', font=('times new roman', 15, 'bold'), bg='white').place(x=20,y=300)
        lbl_sal = Label(emp_frame, text='Emp Salary', font=('times new roman', 15, 'bold'), bg='white').place(x=20,y=340)

        txt_emp_code = Entry(emp_frame,textvariable=self.var_emp_code, font=('times new roman', 15), bg='lightyellow').place(x=200, y=60)
        txt_emp_name = Entry(emp_frame,textvariable=self.var_emp_name, font=('times new roman', 15), bg='lightyellow').place(x=200, y=100)
        txt_department = Entry(emp_frame,textvariable=self.var_emp_department, font=('times new roman', 15), bg='lightyellow').place(x=200, y=140)
        txt_designation = Entry(emp_frame,textvariable=self.var_emp_designation, font=('times new roman', 15), bg='lightyellow').place(x=200, y=180)
        txt_emp_email = Entry(emp_frame,textvariable=self.var_emp_email, font=('times new roman', 15), bg='lightyellow').place(x=200, y=220)
        txt_addr = Entry(emp_frame,textvariable=self.var_emp_addr, font=('times new roman', 15), bg='lightyellow').place(x=200, y=300)
        txt_sal = Entry(emp_frame,textvariable=self.var_emp_sal, font=('times new roman', 15), bg='lightyellow').place(x=200, y=340)

        # radio buttons
        self.gender=StringVar()
        rdbtn = Radiobutton(root, text='Male', variable=self.gender, value=1, font=10, bg='lightyellow').place(x=250, y=420,width=80,height=30)
        rdbtn1 = Radiobutton(root, text='Female',variable=self.gender, value=2, font=10, bg='lightyellow').place(x=350, y=420,width=105,height=30)


        btn1 = Button(emp_frame, text='QR Generate',command=self.generate, font=('times new roman', 15, 'bold'), bg='#2196f3', fg='white').place(x=160, y=390, width=140, height=30)

        btn2 = Button(emp_frame, text='Clear',command=self.clear, font=('times new roman', 15, 'bold'), bg='#063246',fg='white').place(x=320, y=390, width=80, height=30)

        self.msg = ' '
        self.lbl_msg = Label(emp_frame, text=self.msg, font=('times new roman', 15), bg='white', fg='green')
        self.lbl_msg.place(x=20, y=440, relwidth=1)

        # ************************* Employee QR code Window **********************************
        qr_frame = Frame(self.root, bd=2, relief=RIDGE, bg='white')
        qr_frame.place(x=600, y=200, width=250, height=380)

        qr_title = Label(qr_frame, text='Employee QR code', font=("Goudy old Style ", 20), bg='#043256',fg='white').place(x=0, y=0, relwidth=1)

        self.qr_code = Label(qr_frame, relief=RIDGE, text='QR code\n not available', font=('times new roman', 15),bg='#3f51b5', fg='white', bd=1)
        self.qr_code.place(x=35, y=100, width=180, height=180)
   
    def clear(self):
        self.var_emp_code.set('')
        self.var_emp_name.set('')
        self.var_emp_department.set('')
        self.var_emp_designation.set('')
        self.var_emp_email.set('')
        self.gender.set('')
        self.var_emp_addr.set('')
        self.var_emp_sal.set('')
        self.msg = ''
        self.lbl_msg.config(text=self.msg)
        self.qr_code.config(image='')

    def generate(self):
            if self.var_emp_designation.get()==''or self.var_emp_code.get()=='' or self.var_emp_name.get()=='' or \
                    self.var_emp_addr.get()=='' or self.gender.get()=='' or self.var_emp_email.get()=='' or \
                    self.var_emp_sal.get()=='' or self.var_emp_department.get()=='':
                self.msg='All Fields are Required!!!'
                self.lbl_msg.config(text=self.msg,fg='red')
            else:
                qr_data=(f"Employee ID: {self.var_emp_code.get()}\nEmployee Name:{self.var_emp_name.get()}\nDesignation:{self.var_emp_designation.get()}\nDepartment:{self.var_emp_department.get()}\nGender:{self.gender.get()}\nAddress:{self.var_emp_addr.get()}\nE-mail:{self.var_emp_email.get()}\nSalary:{self.var_emp_sal.get()}")
                qr_code=qrcode.make(qr_data)
                qr_code=resizeimage.resize_cover(qr_code,[180,180])
                qr_code.save("Employee_QR/Emp_"+str(self.var_emp_code.get()+'.png'))
                # ******************** Qr code update image *************
                self.im=ImageTk.PhotoImage(file="Employee_QR/Emp_"+str(self.var_emp_code.get()+'.png'))
                self.qr_code.config(image=self.im)
                # ****************** updating notification ************
                self.msg = 'QR Generated successfully!!!'
                self.lbl_msg.config(text=self.msg, fg='green')


root = Tk()
obj = Qr_generator(root)
root.mainloop()
```
